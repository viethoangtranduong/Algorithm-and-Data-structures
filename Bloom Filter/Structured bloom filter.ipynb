```python
import string
import random
import math
```

```python
# Generate input

m = int(input("Number of bits: "))
n = int(input("Number of inputs: "))

# generate initial n elements:
# generae randomly a string of lowercase and numbers, with the length randomly between 1 and 10
def create_string():
    size_of_element = random.randint(1,7)
    # all possible characters
    characters = string.digits + string.ascii_lowercase
    output = ''
    for i in range(size_of_element): 
        output += random.choice(characters)
    return output

list_of_input = []
for i in range(n):
    list_of_input.append(create_string())

# create initial bloom filter
bloom_filter = [0]*m

# ideal number of hasj functions:
k = int(m/n*math.log(2)) + 1
```

```python
# data encryption: we want to deal with number
# change string into number
combination = string.digits + string.ascii_lowercase
dictionary = []
for i in combination:
    dictionary.append(i)

def input_to_number(x):
    x = str(x)
    output = 0
    for i in range(len(x)):
        output += dictionary.index(x[i])*(36**i)
    return output
print(input_to_number("zzzzz"))
print(2**30)
```

```python
# find prime for the universal hash
def find_prime():
    # find the smallest prime number > 2*m**2
    p = random.randrange(2**32 + 1, 2**34 + 1, 2)
    while True:           
        if all(p % i != 0 for i in range(3, int((p ** 0.5) + 1), 2)):
            # all the odd number less than of equal sqrt p
            break            
        else: 
            p += 2
    return p

def hash_coefficient(p):
    a = random.randint(1, p)
    b = random.randint(0, p)
    return [a,b]

def create_hash_function():
    #m: bits, n: inputs
    hash_family = []
    for i in range(k):
        p = find_prime()
        coefficient = hash_coefficient(p)
        coefficient.append(p)
        hash_family.append(coefficient)
    return hash_family

hash_functions = create_hash_function()
print(hash_functions)
```

```python
# the location of element after using hash functions
def encryption_variable(x):
    location = []
    for function in hash_functions:
        location.append(((function[0] * x + function[1]) % function[2]) % m)
    return location
        
```

```python
# bloom filter functions

def insert_element(bloom_filter,x):
    new_input = input_to_number(x)
    encryption = encryption_variable(new_input)
    for k in encryption:
        bloom_filter[k] += 1
    return bloom_filter

def delete_element(bloom_filter,x):
    remove_element = input_to_number(x)
    encryption = encryption_variable(remove_element)
    for k in encryption:
        bloom_filter[k] -= 1
    return bloom_filter
        
def check_element(bloom_filter, x):
    check_element = input_to_number(x)
    encryption = encryption_variable(check_element)
    product_hash = 1
    for k in encryption:
        product_hash *= bloom_filter[k]
    if product_hash == 0:
        return False
    else:
        return True
```

```python
# create the state of bloom filter after n inputs
for x in list_of_input:
    insert_element(bloom_filter,x)
```

```python
test_cases = []
for i in range(10000):
    test_cases.append(create_string())

for i in test_cases:
    result = check_element(bloom_filter, i)
    if result == True:
        if i in list_of_input: 
            print("YES - YES")
        else: 
            print("YES - False positive")
    else: 
        print("NO")
```
